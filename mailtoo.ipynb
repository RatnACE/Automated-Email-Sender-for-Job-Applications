```python
import csv

# Define the custom message template
subject_template = "Application for Technology Position at {}"
body_template = """
Dear {},

I hope this message finds you well. My name is {}, and I am writing to express my strong interest in joining {} as a member of your dynamic technology team.

 
I am excited about the possibility of becoming a part of your team and contributing to the continued success of {}. I am eager to discuss how my skills and experiences align with your organization's needs.

Thank you for considering my application. I look forward to the opportunity of further discussions. Please feel free to reach me at {} or connect with me on LinkedIn at {}.

Sincerely,
{}
{}
"""

# Define the input and output file names
input_file = 'C:/Users /Downloads/hr_list.csv'
output_file = 'C:/Users /Downloads/mailtooutput.csv'

# Read the input CSV file and create the output CSV file
with open(input_file, mode='r') as csv_file, open(output_file, mode='w', newline='') as output_csv_file:
    csv_reader = csv.DictReader(csv_file)
    fieldnames = csv_reader.fieldnames + ['subject', 'body', 'mailto_link']
    
    csv_writer = csv.DictWriter(output_csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()
    LinkedIn_Profile_URL="https://www.linkedin.com/in/ratnesh-kumar-ratan/"
    for row in csv_reader:
        # Extract data from the input CSV
        company_name = row[' ']
        full_name = row['name']
        your_name = " "
        email = row['mailid']
        
        # Extract the first name from the full name
        first_name = full_name.split()[0]

        # Generate the subject and body
        subject = subject_template.format(company_name)
        body = body_template.format(first_name, your_name, company_name,
            company_name, company_name, email, LinkedIn_Profile_URL, your_name, LinkedIn_Profile_URL)

        # Create the mailto link
        mailto_link = f'mailto:{email}?subject={subject}&body={body}'

        # Add the subject, body, and mailto link to the row
        row['subject'] = subject
        row['body'] = body
        row['mailto_link'] = mailto_link

        # Write the row to the output CSV
        csv_writer.writerow(row)

print(f'Mailto links, subject, and body have been generated and saved to {output_file}.')
 
```

```python
import csv
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders

body_template = body_template = """
<html>

<body>
    
    Dear {},
<p>
I hope this message finds you well. My name is <strong>{}</strong>, and I am writing to express my strong interest in joining <strong>{}</strong> as a member of your dynamic technology team.
    </p>
 
 
I am excited about the possibility of becoming a part of your team and contributing to the continued success of <strong>{}</strong>. I am eager to discuss how my skills and experiences align with your organization's needs.
<p>
Thank you for considering my application. I look forward to the opportunity of further discussions. Please feel free to reach me at <strong><a href="mailto:ratnesh.bytes@gmail.com">ratnesh.bytes@gmail.com</a></strong> or connect with me on LinkedIn at <strong><a href="{}">{}</a></strong>
</p>
Sincerely,<br><strong>
<a href="{}">Ratnesh Kumar Ratan</a></strong><br>
<strong><a href="{}">{}</a></strong>
</body>
</html>
"""
# Define your SMTP server and credentials
smtp_server = 'smtp.gmail.com'
smtp_port = 587  # Use the appropriate port for your SMTP server
smtp_username = ' '
smtp_password = app_pass
my_mail=" "
LinkedIn_Profile_URL=" "
sended_list=[]
# Read the CSV file with email data
input_file = 'C:/Users/Downloads/mailtooutput2332.csv'
with open(input_file, mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    for row in csv_reader:
        
        to_email = row['mailid']

        # Create the email body with highlighted points
        company_name = row['companyname']
        full_name = row['name']
        your_name = " "
        email = row['mailid']
        subject = f'Regarding technical opportunities for a fresher at {company_name}'
        # Extract the first name from the full name
        first_name = full_name.split()[0]

        # Generate the subject and body
        
        body = body_template.format(first_name, your_name, company_name,company_name, company_name,
                                    LinkedIn_Profile_URL, LinkedIn_Profile_URL, LinkedIn_Profile_URL, LinkedIn_Profile_URL, LinkedIn_Profile_URL)
        #print( subject,body)
        #exit()

        # Create the email message
        message = MIMEMultipart()
        message['From'] = smtp_username
        message['To'] = to_email
        message['Subject'] = subject
        message.attach(MIMEText(body, 'html'))
        
        
        # Connect to the SMTP server, login, and send the email
        try:
            smtp_conn = smtplib.SMTP(smtp_server, smtp_port)
            smtp_conn.starttls()
            smtp_conn.login(smtp_username, smtp_password)
            smtp_conn.sendmail(smtp_username, to_email, message.as_string())
            smtp_conn.quit()
            print(f'Sent email to {to_email} with subject: {subject}')
            sended_list.append(f'Sent email to {full_name} at {to_email} with subject: {subject}')
        except Exception as e:
            print(f'Error sending email to {to_email}: {str(e)}')
```

```python
with open("mailsendtoo2332.txt", "w") as file:
    # Iterate over the list and write each element to the file
    for item in sended_list:
        file.write(item + "\n")
```
